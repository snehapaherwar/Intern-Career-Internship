```python
# Handling missing data

import pandas as pd

missing_values = df.isnull().sum()         #Assuming df is already defined
print("Missing Values:")
print(missing_values)

df['Subscribers'].fillna(df['Subscribers'].mean(), inplace=True)

missing_values_after_imputation = df.isnull().sum()
print("\nMissing Values After Imputation:")
print(missing_values_after_imputation)
```

```python
#outliers
sns.boxplot(x=df['Subscribers'])
plt.show()
```

```python
# Top youtube streamers and top categories

top_Subscribers = df.sort_values(by = 'Subscribers', ascending = False).head(10)
print("\nTop channels by subscribers")
print(top_Subscribers)
```

```python
# Streamers audience by country

import pandas as pd
import matplotlib.pyplot as plt

categorical_columns = ['Country']

for column in categorical_columns:
  value_counts = pd.Series(df[column]).value_counts()
  plt.figure(figsize = (11,6))
  value_counts.plot(kind='bar', color='purple')
  plt.title(f"Subscribers Distribution of {column}")
  plt.xlabel(column.title())
  plt.ylabel('Subscribers')
  plt.xticks(rotation = 75)
  plt.show()
```

```python
categorical_columns = ['Categories', 'Country']
for column in categorical_columns:
  unique_values = df[column].unique()
  num_unique_values = len(unique_values)
  print(f"Exploring column: {column}")
  print(f"Number of unique values: {num_unique_values}")
  print("Unique Values and their Frequencies:")
  for value, count in df[column].value_counts().items():
    print(f" {value}: {count} ({count / len(df) * 100:.2f}%)")
    print("\n")
```

```python
plt.figure(figsize = (11,6))
sns.histplot(df['Subscribers'], bins=20, kde=True)
plt.title('Distribution of Subscribers')
plt.xlabel('Country')
plt.ylabel('Subscribers')
plt.show()
```

```python
#Finding average
average_metrics = df[{'Subscribers', 'Visits', 'Likes','Comments'}].mean()
print(average_metrics)
```

```python
# Exploring the distribution of content categories
category_subscribers = df.groupby('Categories')['Subscribers'].sum()


# Plotting the distribution of content categories by subscribers

plt.figure(figsize=(28,18))
category_subscribers.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Categories by Subscribers')
plt.xlabel('Category')
plt.ylabel('Subscribers')
plt.xticks(rotation = 45)
plt.show()

# Identify categories with the highest number of subscribers
highest_subscribers_categories = category_subscribers.head(5)
print("Categories with the highest number of subscribers:")
print(highest_subscribers_categories)
```

```python
# Above average

average_subscribers = df['Subscribers'].mean()
average_visits = df['Visits'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()

# Filter streamers with above average performance metrics

above_average_streamers = df[(df['Subscribers'] > average_subscribers) &
                            (df['Visits'] > average_visits) &
                             (df['Likes'] > average_likes) &
                             (df['Comments'] > average_comments)]
```

```python
# content recommendations

grouped_by_country = df. groupby('Country').agg(Subscribers=('Subscribers','sum'), Likes=('Likes', 'sum')).reset_index()
print("Total Subscribers and Likes by Country:")
print(grouped_by_country.to_string())
```
